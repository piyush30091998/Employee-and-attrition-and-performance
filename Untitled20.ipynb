<h2>Data Visualization</h2>

```python
sns.countplot(x='BusinessTravel', hue='Attrition', data=df, palette='CMRmap')
plt.show()
```

```python
sns.countplot(x='MaritalStatus', hue='Attrition', data=df, palette='prism_r')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='JobRole', hue='Attrition', data=df, palette='terrain')
plt.xticks(rotation=45)
plt.show()
```

<h4>Some Observations</h4>

1. Gender: Male employees quit more than female employees.<br>
2. Business Travel: The employees who travel rarely are more likely to quit than other employees.<br>
3. Marital Status: Employees who are single tend to quit their jobs more than the married or divorced.<br>
4. Department: Research and Development employees don't quit their jobs as much as the other
departments.<br>
5. Job Role: Sales Executives, Laboratory Technicians and Research Scientists are more likely to quit
than other employees.<br>
6. Education Field: Employees from Life Sciences, Medical and Marketing educational background are
more likely to stay than other employees of different educational background.<br>
7. Over Time: Employees who do over time, quit more.
